```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df = pd.read_csv('data/train.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Month_Year'] = df['Order Date'].dt.to_period('M').astype(str)
```

```python
monthly_sales = df.groupby('Month_Year')['Sales'].sum().reset_index()
fig1 = px.line(monthly_sales, x='Month_Year', y='Sales', title='Monthly Sales Performance')
```

```python
region_sales = df.groupby('Region')['Sales'].sum().reset_index()
fig2 = px.bar(region_sales, x='Region', y='Sales', color='Region', title='Total Sales by Region')
```

```python
cat_sales = df.groupby('Category')['Sales'].sum().reset_index()
fig3 = px.pie(cat_sales, values='Sales', names='Category', hole=0.4, title='Sales by Product Category')
```

```python

fig1.write_html("sales_trend.html")
fig2.write_html("regional_sales.html")
fig3.write_html("category_distribution.html")
```

```python

fig1.show(renderer="browser")
fig2.show(renderer="browser")
fig3.show(renderer="browser")
```
