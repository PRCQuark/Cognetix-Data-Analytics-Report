```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('data/insurance_data.csv')
df['age'] = df['age'].fillna(df['age'].median())
df['region'] = df['region'].fillna(df['region'].mode()[0])
```

```python
df['claim_group'] = df['claim'].apply(lambda x: 'High' if x > df['claim'].median() else 'Low')
```

```python
plt.figure(figsize=(18, 12))
plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='smoker', y='claim', palette='Set2')
plt.title('Impact of Smoking on Claim Amount')
```

```python
plt.subplot(1,1,1)
sns.barplot(data=df, x='region', y='claim', palette='coolwarm')
plt.title('Average Claim by Region')
```

```python
plt.subplot(1,1,1)
sns.scatterplot(data=df, x='age', y='claim', hue='smoker', alpha=0.6)
plt.title('Age and Smoking status vs Claim')
```

```python
plt.subplot(1,1,1)
sns.boxplot(data=df, x='claim_group', y='bmi', palette='viridis')
plt.title('BMI: High vs Low Claim Profiles')
```

```python
plt.tight_layout()
plt.show()
```
